```python
import pandas as pd
import matplotlib.pyplot as plt

# It is assumed that the data.csv file is already in the same directory.
# If not, you can create it with the data provided in the initial prompt.

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('../data/spikes.csv')

# Drop any rows with missing values
df.dropna(inplace=True)

# Create a single figure and axis
plt.figure(figsize=(15, 8))
ax = plt.gca()

# Plot u vs time as a line plot
ax.plot(df['time'], df['u'], label='u (Neural Activity)', color='blue')

# Find the time points where spikes occur for spike1 and spike2
spike1_times = df[df['spike1'] == 1]['time']
spike2_times = df[df['spike2'] == 1]['time']

# Plot vertical lines for each spike occurrence.
# The 'label' is added only for the first line of each spike type to avoid duplicate legend entries.
for time in spike1_times:
    ax.axvline(x=time, color='red', linestyle='--', linewidth=1, label='spike1' if time == spike1_times.iloc[0] else "")
for time in spike2_times:
    ax.axvline(x=time, color='green', linestyle=':', linewidth=1, label='spike2' if time == spike2_times.iloc[0] else "")

# Set plot titles and labels
ax.set_title('Neural Activity and Spikes Over Time')
ax.set_xlabel('Time')
ax.set_ylabel('u (Neural Activity)')

# Create a single legend for all series
ax.legend()

# Add grid for better readability
ax.grid(True)

# Save the plot
plt.savefig('spikes_combined.png')

# Show the plot
plt.show()
```
