```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('../data/spikes.csv')
df.dropna(inplace=True)

# Create a figure and axis
fig, ax1 = plt.subplots(figsize=(15, 8))

# Plot mean membrane potential (u_mean) as a blue line
ax1.plot(df['time'], df['u_mean'], label='Mean Membrane Potential (u)', color='blue')
ax1.set_xlabel('Time')
ax1.set_ylabel('Mean Membrane Potential (u)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create a second y-axis for total spikes
ax2 = ax1.twinx()
ax2.bar(df['time'], df['spikesum'], width=0.8, alpha=0.3, color='red', label='Total Spikes')
ax2.set_ylabel('Total Spikes', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Titles and grid
plt.title('Neural Activity (u) and Total Spikes Over Time')
ax1.grid(True)

# Legends
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper right')

# Save and show
plt.savefig('spikes_combined.png')
plt.show()
```
